## **K-means**

### **Generate Data**

```python
from sklearn.datasets import make_blobs #นำเข้าฟังก์ชั่นเพื่อจำลองข้อมูลที่เป็นกลุ่มก้อน
```

```python
X, y = make_blobs(n_samples=500, centers=3, n_features=2, random_state=6) 
#กำหนดจำนวนตัวอย่าง 500 ให้มี 3 กลุ่ม มี features=2 โดยสุ่ม
```

```python
X.shape #ตรวจสอบขนาด
```

```python
from matplotlib import pyplot as plt #นำเข้าเพื่อใช้วาดกราฟ
plt.scatter(X[:,0],X[:,1])
```

### **Clustering**

**import**

```python
from sklearn.cluster import KMeans #นำเข้าเพราะเราจะทำการจักกลุ่มคัสเตอร์
```

**Define**

```python
kmeans = KMeans(n_clusters=2, random_state=6) #กำหนดตัวแปร
```

**Fit-Predict**

```python
kmeans.fit(X)
```

```python
y_clust = kmeans.fit_predict(X)
```

```python
y_clust
```

**Plot**

```python
plt.scatter(X[:,0],X[:,1], c=y_clust)
```

**3 Cluster**

```python
#Define
kmeans = KMeans(n_clusters=3, random_state=6)
#Fit-Predict
y_clust = kmeans.fit_predict(X)
#plot
plt.scatter(X[:,0],X[:,1], c=y_clust)
```

**4 Cluster**

```python
#Define
kmeans = KMeans(n_clusters=4, random_state=6)
#Fit-Predict
y_clust = kmeans.fit_predict(X)
#plot
plt.scatter(X[:,0],X[:,1], c=y_clust)
```

**5 Cluster**

```python
#Define
kmeans = KMeans(n_clusters=5, random_state=14)
#Fit-Predict
y_clust = kmeans.fit_predict(X)
#plot
plt.scatter(X[:,0],X[:,1], c=y_clust)
```

```python
#Define
kmeans = KMeans(n_clusters=5, random_state=6)
#Fit-Predict
y_clust = kmeans.fit_predict(X)
#plot
plt.scatter(X[:,0],X[:,1], c=y_clust)
```

### **Example Application (Color Quantization)**

```python
import cv2
import os
#เพิ่มฟังก์ชั่นที่ต้องการใช้เพื่อใช้การภาพ การนำเข้าภาพ และการใช้สี
```

```python
img = cv2.imread('/content/รูป.jpeg') #นำเข้าภาพจากไฟล์ที่อัพโหลด
```

```python
img.shape #ตรวจสอบขนาดของภาพ ไม่ควรใหญ่เกินไป
```

```python
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img) #กำหนนดตัวแปร
```

### การนับจำนวนสี

```python
import numpy as np
```

```python
np.unique(np.reshape(RGB_img,(RGB_img.shape[0]*RGB_img.shape[1],3)),axis=0).shape #มี 43,285 สี
```

> **จัดกลุ่มสีให้เหลือ 16 สี**

```python
myquantize = KMeans(n_clusters=16, random_state=6)
```

แปลงข้อมูลให้อยู่ในรูป row-column

```python
img_row = np.reshape(RGB_img,(RGB_img.shape[0]*RGB_img.shape[1],3))
img_row.shape
```

```python
final_color = myquantize.fit_predict(img_row)
```

ใช้ centroid เป็นตัวแทนของสี

```python
centroid = myquantize.cluster_centers_
```

```python
centroid
```

แทนสีคืน

```python
new_img = np.zeros(RGB_img.shape)
for i in range(RGB_img.shape[0]):
    for j in range(RGB_img.shape[1]):
        new_img[i,j,:] = centroid[myquantize.predict([RGB_img[i,j,:]]),:]
```

```python
plt.imshow(new_img/255)
```

```python
plt.imsave('img.jpg',new_img/255)
```
